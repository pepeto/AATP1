```python
# Carga de librerías.
import pandas as pd
import seaborn as sns; sns.set()
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
%matplotlib inline
```

Lectura de la Base de Datos en un Dataframe de Pandas

```python
# En el nombre del archivo se puede poner alternativamente el path donde se encuentre localmente almacenado.
df = pd.read_csv('healthcare-dataset-stroke-data.csv')
print(df.head)
```

# **Análisis Preliminar de Datos**


Tabla de correlación cruzada para identificar atributos redundante. En este caso no se ven atributos numéricos con alto coeficiente de correlación entre sí.

```python
df.corr(method='pearson')
```

PairPlot para explorar los datos visualmente e identificar correlaciones y patrones

```python
sns.pairplot(df, hue='stroke', size=1.5);
```

```python
# Extracción de los atributos y la clase target
X = df.drop('stroke', axis=1) # Matriz de atributos
y = df['stroke']              # Vector con la columna stroke
```

```python
# Separa un 80% para train y un 20% para test.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
